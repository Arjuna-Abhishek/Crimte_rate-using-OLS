```python
import pandas as pd
f = open(r"C:\Users\abhis\poverty_new_new.csv")
df=pd.read_csv('poverty_new_new.csv', sep=',',header=0)
df[["%jobless","Murders_M"]]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x=df['%jobless']
y=df['Murders_M']
plt.figure(figsize=(12,6))
ax=plt.axes()
ax.set_xlabel("Jobless")
ax.set_ylabel("Murders")
ax.scatter(x,y)
```

```python
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
x=df[['%jobless']].to_numpy()
y=df[['Murders_M']].to_numpy()
plt.figure(figsize=(12,6))
ax=plt.axes()
ax.set_xlabel("Jobless")
ax.set_ylabel("Murders")
ax.scatter(x,y)
model= LinearRegression()
model.fit(x,y)
```

```python
df.describe()
```

```python
model.intercept_
model.coef_
```

#The Coefficient is 6.69275159

```python
model.score(x,y)
```

R-Square is 0.6834326423296082

```python
y_est = model.predict(x)
ax=plt.axes()
ax.set_xlabel("Jobless")
ax.set_ylabel("Murders")
ax.scatter(x,y)
ax.plot(x,y_est)
```

```python
x_new =np.asarray(6)
y_pred=model.predict(x_new.reshape(1,-1))
print("Estimated murders:",y_pred)
y_pred
```

```python
ax=plt.axes()
ax.set_xlabel("Jobless")
ax.set_ylabel("Murders")
ax.scatter(x,y)
ax.plot(x,y_est)
ax.scatter(x_new,y_pred,color='black')
```

```python
import statsmodels.api as sm
X=df[['%jobless']].to_numpy()
Y=df[['Murders_M']].to_numpy()
X=sm.add_constant(X)
model=sm.OLS(Y,X)
results=model.fit()
print(results.summary())
```
